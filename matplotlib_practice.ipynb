```python
# Import matplotlib dependency, from matplotlib import pyplot.
import matplotlib.pyplot as plt
# Import additional dependencies
import random
import numpy as np
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
```

```python
## Create a Line Chart Using the Object-Oriented Interface. fig, ax = plt.subplots()
# Create the plot with ax.plt(), Option 1.
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(), Option 2.
# Use fig = plt.subplots(figsize=(w,h)) to change figure-level attributes (such as axis labels, a title, or a legend) or save the figure as an image.
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt(), Option 3.
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Use plt.show() to look for all the active figure objects and open a window that displays them if there are two or more sets of data to plot.
# Ex: Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Skill Drill
# Create a Scatter Plot Using the Object-Oriented Interface.
fig, ax = plt.subplots()
# Add the s parameter.
# Switch the axis so that the Fare($) data is on the x-axis.
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
ax.scatter(y_axis, 
           x_axis,
           facecolor="skyblue",
           s = [i * 5 for i in y_axis],
           alpha=0.2,
           edgecolor="black",
           lw=2, 
           label="Boston")
# Add a title.
ax.set_title("PyBer Fare by Month")
#Add a legend for the city of Boston.
ax.legend()
# Add axes labels.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Add a limit for the x-axis data.
plt.xlim(0, 50)
# Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
plt.show()
```

```python
# Create Pie Charts.
# Create a Pie Chart Using the MATLAB Method.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign variable with 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# The "explode" parameter offsets the indicated wedge by a fraction of the radius, where "0" is zero distance from the center of the pie, and "1" is completely outside the diameter of the pie.
# Assign tuple variable, explode_values, The length of the tuple should be equal to the number of wedges in the pie chart.
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Increase the size of the pie chart.
plt.subplots(figsize=(8, 8))
# Add the percentage of each wedge on the pie chart, use the autopct parameter and provide the format of one decimal place using .1f%, % before and after the .1f% formats the number as a percentage.
# Add colors= parameter.
plt.pie(y_axis, explode=explode_values, labels=x_axis, colors=colors, autopct='%.1f%%')
plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface.
# Increase the figure size to 8x8.
fig, ax = plt.subplots(figsize=(8, 8))
# Add new colors of your own choosing or use the colors from the previous pie chart.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Explode the two highest percentage months.
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Add a percentage to one decimal place to each wedge of the pie.
# Add exploded values, explode=explode_values.
# Add a shadow, shadow=True.
# Add a start angle so that January is at the top, startangle=90.
# Reverse the order so that the month order is in a clockwise direction, counterclock=False.
# # Add new colors, colors=colors.
plt.pie(y_axis, labels=x_axis, explode=explode_values, shadow=True, startangle=90, counterclock=False, colors=colors, autopct='%.1f%%')
plt.show()
```
